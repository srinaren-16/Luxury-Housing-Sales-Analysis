```python
# Remove duplicate rows
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
# Strip whitespace from column names
df.columns = df.columns.str.strip()
```

```python
# Unique value
for col in df.columns:
    print(f"\n{col}:")
    print(df[col].unique())
```

```python
# Standard categorical columns
text_cols = [
    "Micro_Market",
    "Developer_Name",
    "Transaction_Type",
    "Buyer_Type",
    "Sales_Channel",
    "Possession_Status"
]

for col in text_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.title()
    )
```

```python
# Cleaning Configuration
df["Configuration"] = (
    df["Configuration"]
    .astype(str)
    .str.strip()
    .str.upper()
    .str.replace(" ", "", regex=False)
)
```

```python
# Remove currency symbols and text from Ticket_Price_Cr
df["Ticket_Price_Cr"] = (
    df["Ticket_Price_Cr"]
    .astype(str)
    .str.replace("â‚¹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace("cr", "", case=False, regex=True)
    .str.strip()
)
```

```python
# Convert Ticket_Price_Cr to numeric
df['Ticket_Price_Cr'] = pd.to_numeric(
    df['Ticket_Price_Cr'],
    errors='coerce'
)
```

```python
# Fill missing values for Ticket_Price_Cr
df['Ticket_Price_Cr'] = (
    df['Ticket_Price_Cr']
        .fillna(
            df.groupby(['Micro_Market', 'Configuration'])['Ticket_Price_Cr']
              .transform('median')
        )
        .fillna(df['Ticket_Price_Cr'].median())
)
```

```python
# Fill missing values for Unit_Size_Sqft
df['Unit_Size_Sqft'] = (
    df['Unit_Size_Sqft']
        .fillna(df.groupby('Configuration')['Unit_Size_Sqft'].transform('median'))
        .fillna(df['Unit_Size_Sqft'].median())
)
```

```python
# Fill missing values for Amenity_Score
df['Amenity_Score'] = (
    df['Amenity_Score']
        .fillna(df.groupby('Configuration')['Amenity_Score'].transform('median'))
        .fillna(df['Amenity_Score'].median())
)
```

```python
# Fix Buyer_Comments 
df['Buyer_Comments'] = df['Buyer_Comments'].fillna('No Comments')
```

```python
# Convert Purchase_Quarter to datetime
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
```

```python
# Feature Engineering
```

```python
# Create time features
df['Year'] = df['Purchase_Quarter'].dt.year
df['Quarter'] = df['Purchase_Quarter'].dt.to_period('Q').astype(str)
df['Quarter_Number'] = df['Purchase_Quarter'].dt.quarter
```

```python
# Quick check
df[['Year','Quarter','Quarter_Number']].head()
```

```python
# Price_per_Sqft
df['Price_per_Sqft'] = (df['Ticket_Price_Cr'] * 1e7) / df['Unit_Size_Sqft']
```

```python
# Quick check
df[['Ticket_Price_Cr','Unit_Size_Sqft','Price_per_Sqft']].head()
```

```python
df['Transaction_Type'].unique()
```

```python
# Create booking flag based on transaction type
df["Booking_Flag"] = df["Transaction_Type"].map(
    {"Primary": 1, "Secondary": 0}
)
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:srinaren@localhost:3306/realestate"
)

try:
    engine.connect()
    print("SQLAlchemy engine connected")
except:
    print("Connection failed")
```

```python
from sqlalchemy import text

create_table_query = """
CREATE TABLE IF NOT EXISTS luxury_housing_bangalore (
    Property_ID VARCHAR(20) PRIMARY KEY,
    Micro_Market VARCHAR(100),
    Project_Name VARCHAR(100),
    Developer_Name VARCHAR(100),
    Unit_Size_Sqft FLOAT,
    Configuration VARCHAR(20),
    Ticket_Price_Cr FLOAT,
    Price_per_Sqft FLOAT,
    Transaction_Type VARCHAR(20),
    Booking_Flag INT,
    Buyer_Type VARCHAR(50),
    NRI_Buyer VARCHAR(10),
    Purchase_Quarter DATE,
    Year INT,
    Quarter VARCHAR(10),
    Quarter_Number INT,
    Connectivity_Score FLOAT,
    Amenity_Score FLOAT,
    Locality_Infra_Score FLOAT,
    Avg_Traffic_Time_Min INT,
    Possession_Status VARCHAR(50),
    Sales_Channel VARCHAR(50),
    Buyer_Comments TEXT
);
"""

with engine.connect() as conn:
    conn.execute(text(create_table_query))

print("Table created")
```

```python
# Inserting the data
df.to_sql(
    name='luxury_housing_bangalore',
    con=engine,
    if_exists='append',
    index=False,
    chunksize=5000
)
```

```python
# Total Record Count
query = """
SELECT COUNT(*) AS total_records
FROM luxury_housing_bangalore;
"""

pd.read_sql(query, engine)
```

```python
# Booking Distribution
query_booking = """
SELECT Booking_Flag, COUNT(*) AS total_transactions
FROM luxury_housing_bangalore
GROUP BY Booking_Flag;
"""

pd.read_sql(query_booking, engine)
```

```python
# Builder Performance
query_builder = """
SELECT
Developer_Name,
ROUND(SUM(Ticket_Price_Cr), 2) AS total_revenue_cr,
ROUND(AVG(Ticket_Price_Cr), 2) AS avg_ticket_price_cr
FROM luxury_housing_bangalore
GROUP BY Developer_Name
ORDER BY total_revenue_cr DESC;
"""

pd.read_sql(query_builder, engine)
```

```python
# Micro-Market Conversion
query_market = """
SELECT
    Micro_Market,
    ROUND(AVG(Booking_Flag) * 100,2) AS conversion_rate
FROM luxury_housing_bangalore
GROUP BY Micro_Market
ORDER BY conversion_rate DESC;
"""

pd.read_sql(query_market, engine)
```

```python
# Quarterly Trend
query_quarter = """
SELECT
Quarter,
SUM(Booking_Flag) AS total_bookings,
ROUND(SUM(Ticket_Price_Cr), 2) AS total_revenue_cr
FROM luxury_housing_bangalore
GROUP BY Quarter
ORDER BY Quarter;
"""

pd.read_sql(query_quarter, engine)
```
